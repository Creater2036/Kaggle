```python
import plotly.graph_objs as go
from plotly.offline import plot
import plotly.express as px
```

```python
df = pd.read_csv('/kaggle/input/suicide-rates-overview-1985-to-2016/master.csv')
df.head()
```

Alright so we already have the country and year, we don't really need the country-year column

```python
df.drop('country-year', axis = 1, inplace = True)
```

Before we start visualizing anything lets see how diverse of data we have

```python
df['country'].value_counts() #101 different Countries
```

```python
df['sex'].value_counts() #An equal amount of males and females
```

```python
df['age'].value_counts()
```

```python
df['generation'].value_counts() #This could come useful for group by generation instead of age or gender
```

```python
df_gen = df.groupby(by = 'age')['suicides_no'].sum() #Group all the data by its generation
fig = px.bar(df_gen)
fig.show()
```

Suicides seem to be mainly from people who are in the middle of their life which makes sense but I would've expected more from older people

```python
df_sex = df.groupby('sex')['suicides_no'].sum()
fig = px.bar(df_sex)
fig.show()
```

An incredibly higher suicide rate among males then females
